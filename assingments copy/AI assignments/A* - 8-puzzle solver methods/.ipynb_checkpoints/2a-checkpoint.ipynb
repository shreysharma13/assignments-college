```python
import string
import random


def getString():
 num= '12345678'
 ran = ''.join(random.sample(num,8))
 return ran

def inversion_test(sample):
  sum = 0
  for i in range(len(sample)):
    count = 0
    for j in range(i+1,len(sample)):
      if sample[i] > sample[j]:
        count+=1
    sum = sum + count
  if sum % 2 == 0:
   return True
  else:
   return False



def state_generator():
 sample = getString()
 sample2 = list(sample)
 if inversion_test(sample):
  sample2.insert(random.randint(0,7),"_")
  return sample2
 else:
  return state_generator()



class Node:
    def __init__(self,data,level,festimate):
        """ Initialize the node with the data, level of the node and the calculated festimateue """
        self.data = data
        self.level = level
        self.festimate = festimate
    def generate_child(self):
        """ Generate child nodes from the given node by moving the blank space
            either in the four directions {up,down,left,right} """
        x,y = self.find(self.data,'_')
        """ val_list contains position values for moving the blank space in either of
            the 4 directions [up,down,left,right] respectively. """
        val_list = [[x,y-1],[x,y+1],[x-1,y],[x+1,y]]
        children = []
        for i in val_list:
            child = self.shuffle(self.data,x,y,i[0],i[1])
            if child is not None:
                child_node = Node(child,self.level+1,0)
                children.append(child_node)
        return children
    def find(self,puz,x):
        """ Specifically used to find the position of the blank space """
        for i in range(0,len(self.data)):
            for j in range(0,len(self.data)):
                if puz[i][j] == x:
                    return i,j
    def shuffle(self,puz,x1,y1,x2,y2):
        """ Move the blank space in the given direction and if the position value are out
            of limits the return None """
        if x2 >= 0 and x2 < len(self.data) and y2 >= 0 and y2 < len(self.data):
            puz_temp = []
            puz_temp = self.copy(puz)
            temp = puz_temp[x2][y2]
            puz_temp[x2][y2] = puz_temp[x1][y1]
            puz_temp[x1][y1] = temp
            return puz_temp
        else:
            return None
    def copy(self,root):
        """ Copy function to create a similar matrix of the given node"""
        temp = []
        for i in root:
            t = []
            for j in i:
                t.append(j)
            temp.append(t)
        return temp

    


class Puzzle:
    def __init__(self,size):
        """ Initialize the puzzle size by the specified size,open and closed lists to empty """
        self.n = size
        self.open = []
        self.closed = []
    def f(self,start,goal):
        """ Heuristic Function to calculate hueristic value f(x) = h(x) + g(x) """
        return self.h(start.data,goal)+start.level
    def h(self,start,goal):
        """ Calculates the different between the given puzzles """
        temp = 0
        for i in range(0,self.n):
            for j in range(0,self.n):
                if start[i][j] != goal[i][j] and start[i][j] != '_':
                    temp += 1
        return temp
    def get_input(self):
        """ get_inputs the puzzle from the user """
        puz = []
        for i in range(0,self.n):
            temp = input().split(" ")
            puz.append(temp)
        return puz
    def solve(self):
        move = 0
        """ get_input Start and Goal Puzzle state"""
        print("Enter the start state matrix \n")
        start = self.get_input()
        print("Enter the goal state matrix \n")
        goal = [['1','2','3'],['4','5','6'],['7','8','_']]

        start = Node(start,0,0)
        start.festimate = self.f(start,goal)
        """ insert start in open"""

        self.open.append(start)
        print("\n\n")
        while True:
            cur = self.open[0]
            print("")
            print("  | ")
            print("  | ")
            print(" \\\'/ \n")
            print(cur.data)
            for i in cur.data:
                for j in i:
                    print(j,end=" ")
                print("")
            move +=1
            print('no. of moves',move)
            """ checks if state and goal are the same"""
            if(self.h(cur.data,goal) == 0):
                break
            for i in cur.generate_child():
                i.festimate = self.f(i,goal)
                self.open.append(i)
            self.closed.append(cur)
            del self.open[0]

            """ sort open list using festimate """
            self.open.sort(key = lambda x:x.festimate,reverse=False)

state_generator()
puz = Puzzle(3)
puz.solve()
```
